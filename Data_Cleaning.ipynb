```python
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/USER/DA_Projects/NASA-Asteroids-Classification/Asteroids.csv')
```

```python
df.info()
```

```python
#remove unnecessary columns
df.drop('nasa_jpl_url', inplace = True, axis = 1)
df.drop('links.self', inplace = True, axis = 1)
```

```python
#rename columns 
df.columns = ['neo_reference_id','name', 'absolute_magnitude_h', 'hazardous', 'est_dia_min(km)', 'est_dia_max(km)',
              'est_dia_min(m)', 'est_dia_max(m)', 'est_dia_min(miles)', 'est_dia_max(miles)', 'est_dia_min(feet)', 
              'est_dia_max(feet)', 'close_approach_date', 'epoch_date_close_approach', 'relative_velocity(km/s)', 
              'relative_velocity(km/h)', 'relative_velocity(miles/h)', 'miss_distance(astronomical)', 'miss_distance(lunar)',
              'miss_distance(km)', 'miss_distance(miles)', 'orbiting_body', 'orbit_id', 'orbit_determination_date', 'orbit_uncertainty',
              'minimum_orbit_intersection', 'jupiter_tisserand_invariant', 'epoch_osculation', 'eccentricity', 'semi_major_axis', 'inclination',
              'ascending_node_longitude', 'orbital_period', 'perihelion_distance', 'perihelion_argument', 'aphelion_distance', 'perihelion_time',
              'mean_anomaly', 'mean_motion', 'equinox']
```

```python
df
```

```python
#move column 'hazardous' (target) and apeend it to the end.
df = df[[c for c in df if c not in ['hazardous']] 
       + ['hazardous']]
```

```python
df
```

```python
df.columns
```

```python
#transform 'hazardous' labels true and false to the binary 1 and 0. ie, 1 -> true, 0 -> false
df['hazardous'].replace({True : 1, False : 0}, inplace = True)
```

```python
#1. neo_reference_id, name and orbit_id are not needed for our model, since they are used by the researchers for identifying the asteroids or orbit.
#2. few of the columns among are same measures, but in different standards.
    #est_dia_min, est_dia_max, relative_velocity, miss_distance and close_approach_date and epoch_date_close_approach
    #reserve one column from each measure and drop the remainings.
# 3. drop 'orbiting_body' and 'equinox' due to their single value.
#let's drop those columns

df.drop(['neo_reference_id', 'name', 'orbit_id', 'est_dia_min(m)', 'est_dia_max(m)', 'est_dia_min(miles)', 
                'est_dia_max(miles)', 'est_dia_min(feet)', 'est_dia_max(feet)', 'epoch_date_close_approach', 'relative_velocity(km/s)',
                'relative_velocity(miles/h)', 'miss_distance(astronomical)', 'miss_distance(lunar)', 'miss_distance(miles)',
               'orbiting_body', 'equinox'], axis = 1, inplace = True)
```
